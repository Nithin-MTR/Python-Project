# 3. Display the percentage distribution of the data in each year using the pie chart.


```python
import matplotlib.pyplot as plt
year=data['year'].value_counts()
year

plt.pie(data.year.value_counts(), autopct = '%.1f%%', radius = 1.2,labels=year.index)
plt.xlabel('Year')
plt.ylabel('Count')
plt.title('Distribution of data in each year')
plt.show()
```

# 6. Plot and understand the relationship between categorical variable 'year' and a numerical variable 'prodvalue' through boxplot, and write your findings.

```python
sns.boxplot(x='year',y='prodvalue',data=data)
plt.title("relationship between year and prodvalue")
plt.show()
```

```python
#The position of the box along the y-axis provides an understanding of the median 'prodvalue' for each year.
#we can compare the medians across different years to identify changes over time.
```

# 7. Visualize and understand the relationship between the multiple pairs of variables throughout different years using pairplot and add your inferences. (use columns 'numcol', 'yield percol', 'total prod', 'prodvalue','year')

```python
columns=["numcol",'yieldpercol','totalprod','prodvalue','year']
df=data[columns]
sns.pairplot(df,hue="year")
plt.show()
```

```python
#Year-wise variations By using the 'hue' parameter in the pairplot.
#we can observe the relationship between variables for different years. 
```
