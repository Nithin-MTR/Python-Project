# Convert datatype of column published_at from object to pandas datetime.

```python
data['published_at'] = pd.to_datetime(data['published_at'])
data['published_at']
```

```python
# Verify the data type conversion
print("Data types after conversion:")
print(data.dtypes)
```

# Replace the numbers in the column published_month as names of the months i,e., 1 as 'Jan', 2 as 'Feb' and so on.....

```python
data['published_month'] = data['published_at'].dt.month.replace(
    {1: 'Jan', 2: 'Feb', 3: 'Mar', 4: 'Apr', 5: 'May', 6: 'Jun', 7: 'Jul', 8: 'Aug', 9: 'Sep', 10: 'Oct', 11: 'Nov', 12: 'Dec'})

# Verify the changes
print(data['published_month'])
```

```python
#Checks whether the published date has changed to month or not
data
```

# Find the number of videos published each month and arrange the months in a decreasing order based on the video count.

```python
video_count = data.groupby(data['published_month'])['video_id'].count().sort_values(ascending=False)
print(video_count)
```

# Find the count of unique video_id, channel_id and channel_title.

```python
unique_count = data[['video_id', 'channel_id', 'channel_title']].nunique()
print(unique_count)
```

# Find the title of the video which has the maximum number of likes and the title of the video having minimum likes and write your inferences.

```python
top_10_channels=data['channel_title'].value_counts().nlargest(10)
top_10_channels
```

```python
bottom_10_channels=data['channel_title'].value_counts().nsmallest(10)
bottom_10_channels
```
